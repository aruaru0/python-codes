# Pandasのデータフレームを作成する方法

## listから作成する

```python
import pandas as pd

df = pd.DataFrame([[1,0.1,'a'],[3,1.2,'b']], columns=['x', 'y', 'z'])
display(df)
```

```python
df.values, df.columns, df.index
```

## tupleから作成

```python
import pandas as pd

df = pd.DataFrame(((1,0.1,'a'), (3,1.2,'b')), columns=['x', 'y', 'z'])
display(df)
```

## dictから作成する

```python
import pandas as pd

df = pd.DataFrame({'x': [1,3], 'y':[0.1,1.2], 'z':['a', 'b']})
display(df)
```

## numpyから作成する

```python
import pandas as pd
import numpy as np

array = np.array([[1,0.1,'a'],[3,1.2,'b']])
df = pd.DataFrame(array, columns=['x', 'y', 'z'])
display(df)
```

# csvファイルから作成

### ヘッダ付き

```python
%%writefile test.csv
x,y,z
1, 0.1, a
3, 1.2, b
```

```python
import pandas as pd

df = pd.read_csv('test.csv')
display(df)
```

```python
%%writefile test.csv
　　　　x, 　　　　y, 　　　z
1, 0.1, a
3, 1.2, b
```

```python
import pandas as pd

df = pd.read_csv('test.csv')
df.columns
```

### ヘッダなし

```python
%%writefile test.csv
1, 0.1, a
3, 1.2, b
```

```python
import pandas as pd

df = pd.read_csv('test.csv', header=None, names=['x', 'y', 'z'])
display(df)
```

## 行をスキップして読み込み

```python
%%writefile test.csv
# データ列の説明
# x : 整数
# y : 小数点
# z : 文字列
x, y, z
1, 0.1, a
3, 1.2, b
```

```python
import pandas as pd

df = pd.read_csv('test.csv', skiprows=4)
display(df)
```

### スキップする行を指定

```python
%%writefile test.csv
# データ列の説明
# begin data
x, y, z
1, 0.1, a
3, 1.2, b
# end of data
```

```python
import pandas as pd

df = pd.read_csv('test.csv', skiprows=[0,1,5])
display(df)
```

### 先頭のn行を読み込む

```python
%%writefile test.csv
x, y, z
1, 0.1, a
3, 1.2, b
# end of data
a, b
1,2
3,4
```

```python
import pandas as pd

df = pd.read_csv('test.csv', nrows=2) # headerの行は含まれないことに注意
display(df)
```

## データ型を指定した読み込み

```python
%%writefile test.csv
x,y,z
1,2,3
4,5,6
```

```python
import pandas as pd
df = pd.read_csv('test.csv', dtype={'x': int, 'y': float, 'z': str})
display(df)

print(df.dtypes)
```

```python
import pandas as pd

df = pd.read_csv('test.csv', dtype={'x':int, ' y':float, ' z':str})
display(df)

print(df.dtypes)
```

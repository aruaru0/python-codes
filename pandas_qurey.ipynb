```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=10000000, #1000万行のデータを生成
                           n_features=10,
                           n_informative=6,
                           n_classes=5,
                           random_state=42)
```

```python
import pandas as pd
df = pd.DataFrame(X)
df['Y'] = y
df.columns = ['col_X' + str(i) for i in range(1, 11)] + ['col_Y']
df

LOOP = 100
```

```python
import time
start = time.time()  # 現在時刻（処理開始前）を取得

# 実行したい処理を記述
end = time.time()  # 現在時刻（処理完了後）を取得
time_diff = end - start  # 処理完了後の時刻から処理開始前の時刻を減算する
print(time_diff)  # 処理にかかった時間データを使用
```

# 大小比較

```python
start = time.time()

for _ in range(LOOP):
  sel = df[df['col_X1'] > 0]

end = time.time()
time_diff_0 = end - start
print(time_diff_0, "sec.")

display(sel)
```

```python
start = time.time()

for _ in range(LOOP):
  sel = df.query('col_X1 > 0')

end = time.time()
time_diff_1 = end - start
print(time_diff_1, "sec.")

display(sel)
```

```python
val = 0
start = time.time()

for _ in range(LOOP):
  sel = df.query('col_X1 > @val')

end = time.time()
time_diff_2 = end - start
print(time_diff_2, "sec.")

display(sel)
```

```python
import matplotlib.pyplot as plt
import japanize_matplotlib

x = [3,2,1]
y = [time_diff_0, time_diff_1, time_diff_2]

plt.figure(figsize=(12, 3))
plt.barh(x, y)
plt.yticks(x, ["①", "②", "③"])
```

```python
start = time.time()

for _ in range(LOOP):
  sel = df[df['col_Y'].isin([0,1])]

end = time.time()
time_diff_0 = end - start
print(time_diff_0, "sec.")

display(sel)
```

```python
start = time.time()

for _ in range(LOOP):
  sel = df.query('col_Y in [0,1]')

end = time.time()
time_diff_1 = end - start
print(time_diff_1, "sec.")

display(sel)
```

```python
import matplotlib.pyplot as plt
import japanize_matplotlib

x = [2,1]
y = [time_diff_0, time_diff_1]

plt.figure(figsize=(12, 2))
plt.barh(x, y)
plt.yticks(x, ["①", "②"])
```

# 複数条件

```python
start = time.time()

for _ in range(LOOP):
  sel = df[(df['col_X1'] > 0) & (df['col_X2'] < 0) & (df['col_X3'] > 0) & (df['col_X4'] < 0.1) & (df['col_X5'] > 1)]

end = time.time()
time_diff_0 = end - start
print(time_diff_0, "sec.")

display(sel)
```

```python
start = time.time()

for _ in range(LOOP):
  sel = df.query('col_X1 > 0 and col_X2 < 0 and col_X3 > 0 and col_X4 < 0.1 and col_X5 > 1')

end = time.time()
time_diff_1 = end - start
print(time_diff_1, "sec.")

display(sel)
```

```python
import matplotlib.pyplot as plt
import japanize_matplotlib

x = [2,1]
y = [time_diff_0, time_diff_1]

plt.figure(figsize=(12, 2))
plt.barh(x, y)
plt.yticks(x, ["①", "②"])
```

```python
start = time.time()

for _ in range(LOOP):
  sel = df[(df['col_X1'] > 0) | (df['col_X2'] < 0)]

end = time.time()
time_diff_0 = end - start
print(time_diff_0, "sec.")

display(sel)
```

```python
start = time.time()

for _ in range(LOOP):
  sel = df.query('col_X1 > 0 or col_X2 < 0')

end = time.time()
time_diff_1 = end - start
print(time_diff_1, "sec.")

display(sel)
```

```python
import matplotlib.pyplot as plt
import japanize_matplotlib

x = [2,1]
y = [time_diff_0, time_diff_1]

plt.figure(figsize=(12, 2))
plt.barh(x, y)
plt.yticks(x, ["①", "②"])
```
